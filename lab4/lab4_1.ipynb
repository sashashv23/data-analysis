```python
import datetime
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tabulate import tabulate
from datetime import time as dt_time
import time

print("Setup ok")
file = "household_power_consumption.csv"
```

```python
def print_head(data, n=5):
    print_type = "DataFrame" if isinstance(data, pd.DataFrame) else "NumPy" if isinstance(data, np.ndarray) else None
    
    if print_type == "DataFrame":
        table_data = data.head(n)
        print("DataFrame:")
        print(tabulate(table_data.values, headers=table_data.columns, tablefmt='rounded_grid'))
    elif print_type == "NumPy":
        print("\nNumPy:")
        print(tabulate(data[:n], tablefmt='rounded_grid'))
    else:
        print("Incorrect data type")

def create_numpy_arr(file):
    return np.genfromtxt(file, delimiter=',')

def create_data_frame(file):
    column_names = ["Date", "Time", "Global_active_power", "Global_reactive_power", "Voltage", "Global_intensity", "Sub_metering_1", "Sub_metering_2", "Sub_metering_3"]
    column_types = {
        'Global_active_power': float,
        'Global_reactive_power': float,
        'Voltage': float,
        'Global_intensity': float,
        'Sub_metering_1': float,
        'Sub_metering_2': float,
        'Sub_metering_3': float
    }
    
    df = pd.read_csv(file, names=column_names, dtype=column_types, skiprows=1, na_values=['?'])

    df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
    df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S').dt.time

    df.dropna(inplace=True)  
    
    return df

def measure_time(func, *args):
    start = time.time()
    result = func(*args)
    elapsed_time = time.time() - start
    return result, elapsed_time

df = create_data_frame(file)
np_arr = df.to_numpy()
print(np_arr.shape)
print_head(df)
print_head(np_arr)
```

1. Обрати всі домогосподарства, у яких загальна активна споживана

```python
def task1_pd(df):
    task1_df = df[df['Global_active_power'] > 5.0]
    return task1_df

def task1_np(np_arr):
    task1_df_np = np_arr[np_arr[:,2] > 5.0]
    return task1_df_np

task1_df, time_pd = measure_time(task1_pd, df)
task1_df_np, time_np = measure_time(task1_np, np_arr)

print("\nTask 1:")
print(f"Time:\nPd: {time_pd:.6f} seconds \nNp: {time_np:.6f} seconds")
print_head(task1_df)
print_head(task1_df_np)
```

2. Обрати всі домогосподарства, у яких вольтаж перевищую 235 В.

```python
def task2_pd(df):
    task2_df = df[df['Voltage'] > 235]
    return task2_df

def task2_np(np_arr):
    task2_df_np = np_arr[np_arr[:, 4] > 235]
    return task2_df_np

task2_df, time_pd = measure_time(task2_pd, df)
task2_df_np, time_np = measure_time(task2_np, np_arr)

print("\nTask 2:")
print(f"Time:\nPd: {time_pd:.6f} seconds \nNp: {time_np:.6f} seconds")
print_head(task2_df)
print_head(task2_df_np)
```

3. Обрати всі домогосподарства, у яких сила струму лежить в межах 19-20 А, для них виявити ті, у яких пральна машина та холодильних споживають більше, ніж бойлер та кондиціонер.

```python
def task3_pd(df):
    task3_df = df[(df['Global_intensity'] >= 19) & (df['Global_intensity'] <= 20)]
    task3_df = task3_df[(task3_df['Sub_metering_2'] > task3_df['Sub_metering_3']) &
                              (task3_df['Sub_metering_2'] > task3_df['Sub_metering_1'])]
    return task3_df

def task3_np(np_arr):
    task3_df = np_arr[(np_arr[:, 5] >= 19) & (np_arr[:, 5] <= 20)]
    task3_df = task3_df[(task3_df[:, 7] > task3_df[:, 8]) & 
                              (task3_df[:, 7] > task3_df[:, 6])]
    return task3_df

task3_df_pd, time_pd = measure_time(task3_pd, df)
task3_df_np, time_np = measure_time(task3_np, np_arr)

print("\nTask 3:")
print(f"Time:\nPd: {time_pd:.6f} seconds \nNp: {time_np:.6f} seconds")
print_head(task3_df_pd)
print_head(task3_df_np)
```

4. Обрати випадковим чином 500000 домогосподарств (без повторів
елементів вибірки), для них обчислити середні величини усіх 3-х
груп споживання електричної енергії, а також

```python
def task4_pd(df, n=500000):
    sampled_df = df[['Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']].sample(n, replace=True)
    return sampled_df.mean()

def task4_np(np_arr, n=500000):
    sampled_np = np_arr[np.random.choice(np_arr.shape[0], n, replace=True), 6:9]
    return np.mean(sampled_np, axis=0)

averages_pd, time_pd = measure_time(task4_pd, df, n=500000)
averages_np, time_np = measure_time(task4_np, np_arr, n=500000)

print("\nTask 4")
print(f"Averages (Pandas):", *averages_pd.values)  
print(f"Time: {time_pd:.6f} seconds")
print(f"Averages (NumPy):", *averages_np) 
print(f"Time: {time_np:.6f} seconds")
```

5. Обрати ті домогосподарства, які після 18-00 споживають понад 6
кВт за хвилину в середньому, серед відібраних визначити ті, у яких
основне споживання електроенергії у вказаний проміжок часу
припадає на пральну машину, сушарку, холодильник та освітлення
(група 2 є найбільшою), а потім обрати кожен третій результат із
першої половини та кожен четвертий результат із другої половини.

```python
def task5_pd(df):
    filtered_df = df[(df['Time'] > pd.to_datetime('18:00:00').time()) & 
                     (df['Global_active_power'] > 6) & 
                     (df['Sub_metering_2'] > df['Sub_metering_1']) & 
                     (df['Sub_metering_2'] > df['Sub_metering_3'])]

    fh = filtered_df.iloc[:len(filtered_df) // 2]
    sh = filtered_df.iloc[len(filtered_df) // 2:]
    result = pd.concat([fh.iloc[2::3], sh.iloc[3::4]])

    return result

def task5_np(np_arr):
    condition = (np_arr[:, 1].astype('O').astype(dt_time) >= dt_time(18, 0, 0)) & \
                (np_arr[:, 2] > 6) & \
                (np_arr[:, 7] > np_arr[:, 6]) & \
                (np_arr[:, 7] > np_arr[:, 8])
    
    filtered_array = np_arr[condition]
    fh = filtered_array[:len(filtered_array) // 2]
    sh = filtered_array[len(filtered_array) // 2:]
    result = np.concatenate([fh[2::3], sh[3::4]])
    
    return result

filtered_df, time_pd = measure_time(task5_pd, df)
filtered_np, time_np = measure_time(task5_np, np_arr)

print("\nTask 5:")
print(f"Time:\nPd: {time_pd:.6f} seconds \nNp: {time_np:.6f} seconds")
print_head(filtered_df)
print_head(filtered_np)
```
